```python
types = ['Fibers', 'Fragments', 'Films', 'Microbeads', 'Foams']
percentages = [40, 25, 15, 10, 10]

# Create a pie chart
plt.figure(figsize=(7, 7))
plt.pie(percentages, labels=types, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange', 'purple'], startangle=140)
plt.title('Ocean Microplastic Composition')
plt.show()
```

```python
import pandas as pd

# Sample data: Presence of different types of microplastics
data = {
    'Microplastic Type': ['Fibers', 'Fragments', 'Films', 'Microbeads', 'Foams'],
    'Presence': ['Yes', 'Yes', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)
print(df)
```

```python
# Visualizing microplastic concentration distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Microplastic_Concentration'], bins=20, kde=True, color='blue')
plt.title('Distribution of Microplastic Concentration in Water')
plt.xlabel('Microplastic Concentration (Particles/Liter)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of microplastic concentration by location
plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], c=df['Microplastic_Concentration'], cmap='Reds', alpha=0.7)
plt.colorbar(label='Microplastic Concentration (Particles/Liter)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Microplastic Concentration')
plt.show()
```
